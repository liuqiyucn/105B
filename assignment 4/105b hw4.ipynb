```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
start = 0
prob = 0.5

def random_walk(start, p, step):
    out = []
    for i in range(step):
        out.append(start)
        start = start + np.random.choice(np.array([1,-1]), p=[prob, 1-prob])
    return out

output1 = random_walk(0, 0.5, 1000)
output2 = random_walk(0, 0.5, 1000)
output3 = random_walk(0, 0.5, 1000)

time = np.arange(1000)
plt.plot(time, output1)
plt.plot(time, output2)
plt.plot(time, output3)
plt.show()
```

```python
def random_walk_gas(start, step):
    out = []
    for i in range(step):
        out.append(start)
        start = start + np.random.randn()
    return out

output1 = random_walk_gas(0, 1000)
output2 = random_walk_gas(0, 1000)
output3 = random_walk_gas(0, 1000)

time = np.arange(1000)
plt.plot(time, output1)
plt.plot(time, output2)
plt.plot(time, output3)
plt.show()
```

```python
array1 = []
array2 = []
for i in range(10000):
    output_bio = random_walk(0, 0.5, 1000)
    output_gas = random_walk_gas(0, 0.5, 1000)
    array1.append(output_bio[-1])
    array2.append(output_gas[-1])

array1 = np.array(array1)
array2 = np.array(array2)

```

```python
plt.hist(array1, label = "fix")
plt.legend()
plt.show()
plt.hist(array2, label = 'gas')
plt.legend()
plt.show()
```

$$ E(x) = N(2p-1) = N*(x_0), x_0 = 0.1$$
$$ E(\sigma) = 4Np(1-p) = N \sigma^2, \sigma = 0.995$$

```python
array1 = []
array2 = []
for i in range(10000):
    output_bio = random_walk(0, 0.55, 1000)
    output_gas = random_walk_gas(0, 0.5, 1000)
    array1.append(np.abs(output_bio[-1]))
    array2.append(np.abs(output_gas[-1]))

array1 = np.array(array1)
array2 = np.array(array2)

plt.hist(array1, bins=100)
plt.hist(array2, bins=100)
```
