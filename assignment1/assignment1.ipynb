```python
def add_array(a,b):
    #print(a.shape, b.shape)
    if a.shape != b.shape:
        return "Dimension Error"
    else:
        return a+b

array_1 = np.array([3,5,7])
array_2 = np.array([1,1,1])
array_3 = np.array([2,2,2,2])

print(add_array(array_1,array_2))
print(add_array(array_1,array_3))
```
